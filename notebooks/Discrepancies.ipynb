```python
human = {re.search("_(chr[\dYXM]*)\.extended", f.name).group(1): f for f in Path("chm13").glob("*.txt")}
human
```

```python
from tqdm import tqdm

total = 0
total_max = 0
total_chr = {}
for c in tqdm(human):
    temp = pd.read_csv(human[c], 
                       header=None, 
                       names=["seqID", "start", "end", "sequence", "length", "type"]).query("length > 60")
    total += temp.shape[0]
    c_all = temp.groupby(["seqID", "end"], as_index=False)\
                          .agg({"start": "min", "length": "max"}).shape[0]
    total_max += c_all
    total_chr.update({c: c_all})
total, total_max
```
